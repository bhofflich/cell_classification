```python
valid = ct.unit_table.query('valid == True and run_id == "000"')
```

```python
cch_delays = np.array(list(valid.iloc[0]['CCH_delays'].a))
cell_dists = np.zeros((len(valid), len(valid)))
cell_area_overlaps = np.zeros((len(valid), len(valid)))
cell_overlaps = np.zeros((len(valid), len(valid)))
cch_1ms = np.zeros((len(valid), len(valid), len(cch_delays)))
cch_10ms = np.zeros((len(valid), len(valid), len(cch_delays)))
cell_types = np.empty(len(valid), dtype=object)
for i, (uid, cell) in enumerate(valid.iterrows()):
    cell_dists[i, :] = np.array(list(cell.rf_overlaps.a.values()))
    cell_area_overlaps[i, :] = np.array(list(cell.rf_area_overlaps.a.values()))
    cell_overlaps[i, :] = np.array(list(cell.rf_inner_products.a.values()))
    cell_types[i] = cell.label_manual_text
    cch_1ms[i, :, :] = np.array(list(cell['1ms CCHs'].a.values()))
    cch_10ms[i, :, :] = np.array(list(cell['10ms CCHs'].a.values()))
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Reshape the 3D matrix into a 2D matrix where each row is an observation
n = cch_10ms.shape[0]
d = cch_10ms.shape[2]
mask = (cell_dists < 3) & (cell_dists > 0.05)
flat_cch_10ms = cch_10ms[mask]
observations, _ = flat_cch_10ms.shape
flat_cch_10ms = flat_cch_10ms.reshape((observations, d))

# Standardize the features
scaler = StandardScaler()
flat_cch_10ms_std = scaler.fit_transform(flat_cch_10ms)

# Apply PCA
pca = PCA(n_components=d)  # Reduce to 2 dimensions for the sake of example
matrix_2d_pca = pca.fit_transform(flat_cch_10ms_std)

# The explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance_ratio)

# The cumulative explained variance ratio
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print("Cumulative explained variance ratio:", cumulative_explained_variance)

# To decide how many components are needed to capture significant variance
# you might want to look at how many components explain, for example, at least 95% of the variance
n_components = np.argmax(cumulative_explained_variance >= 0.95) + 1
print(f"Number of components to capture 95% variance: {n_components}")
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Reshape the 3D matrix into a 2D matrix where each row is an observation
n = cch_10ms.shape[0]
d = cch_10ms.shape[2]
labels = np.array([[[cell_types[i], cell_types[j]] for i in range(n)] for j in range(n)])
mask = (cell_dists < 1.5) & (cell_dists > 0.05) & ((labels == np.array(['ON parasol','ON parasol'])).all(axis=2))
flat_cch_10ms = cch_10ms[mask]
observations, _ = flat_cch_10ms.shape
flat_cch_10ms = flat_cch_10ms.reshape((observations, d))
#normalize
flat_cch_10ms = flat_cch_10ms / np.linalg.norm(flat_cch_10ms, axis=1)[:, None]


# Standardize the features
# scaler = StandardScaler()
# flat_cch_10ms_std = scaler.fit_transform(flat_cch_10ms)

# Apply PCA
pca = PCA(n_components=25)  # Reduce to 2 dimensions for the sake of example
matrix_2d_pca = pca.fit_transform(flat_cch_10ms)

# The explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance_ratio)
print("Explained variance: ", np.sum(explained_variance_ratio))
```

```python
reduced_cchs = np.matmul(matrix_2d_pca,pca.components_)
#step through plot indices with user input
for i in range(0, len(cch_delays)):
    plt.plot(cch_delays, (flat_cch_10ms[i, :])/np.linalg.norm(flat_cch_10ms[i, :]))
    plt.plot(cch_delays, (reduced_cchs[i, :])/np.linalg.norm(reduced_cchs[i, :]))
    plt.show()
    inp = input("Press Enter to continue...")
    if inp == 'q':
        break
    plt.close()
```
